<h1>Import the data set</h1>

First, let's import pandas (for visualization) and read the csv file for the data set of arrests in the Pittsburgh area

```python
import pandas as pd

arrests = "https://data.wprdc.org/datastore/dump/e419c20c-8df4-4729-830c-e49427a656e0"
arrest_data = pd.read_csv(arrests)
arrest_data.head()
```

<h1>Rank the neighborhoods</h1>

Then we can sort and rank the neighborhoods by the number of crimes in each neighborhood

```python
crimes_by_neighborhood = arrest_data['Neighborhood'].value_counts().reset_index()

print("Crimes by Neighborhood:")
print(crimes_by_neighborhood.to_string())
```

<h1>Number of each type of charge in each neighborhood</h1>

For a little extra data we can sort the number of charges in each neighborhood to see the level of severity of the neighborhoods crimes

```python
arrest_data['Offense_Category'] = arrest_data['ArrestCharge_Felony_Misdemeanor_Description'].str.split(' - ').str[0]

crimes_by_neighborhood_and_offense = arrest_data.groupby(['Neighborhood', 'Offense_Category']).size().reset_index(name='Number of Arrests')

pivot_table = arrest_data.pivot_table(
    index='Neighborhood', 
    columns='Offense_Category',
    values='ArrestCharge_Felony_Misdemeanor_Description',
    aggfunc='count',
    fill_value=0
)

print("Neighborhoods (rows) by Offense Category (columns):")
print(pivot_table.to_string())
```
